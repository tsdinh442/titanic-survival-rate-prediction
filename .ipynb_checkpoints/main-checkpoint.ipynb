```python
sns.countplot(x=train_data['Survived'])
plt.show()
```

- <b> Display data based on sex </b>

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

sns.countplot(x=train_data['Sex'], ax=axes[0])
sns.countplot(x=train_data['Survived'], hue='Sex', data=train_data, ax=axes[1])

plt.tight_layout()
plt.show()
```

- <b> Display data based on ticket class </b>

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

sns.countplot(x=train_data['Pclass'], ax=axes[0])
sns.countplot(x=train_data['Pclass'], hue='Survived', data=train_data, ax=axes[1])

plt.tight_layout()
plt.show()
```

Interpretation: Despite the number of third class passengers is the highest, they are less likely to survive compared to the uppper classes. First class passengers are the most likely to survive. 

- <b> Display data based on the number of siblings/spouses </b>

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

sns.countplot(x=train_data['SibSp'], ax=axes[0])

sns.countplot(x=train_data['SibSp'], hue='Survived', data=train_data, ax=axes[1])

plt.tight_layout()
plt.show()                        
```
